### Modelo Conceptual para calcula la conducción de calor en 1D

Se utiliza la Ley de Fourier para la conducción de calor

1. Consideramos que no hay flujo de calor en las paredes horizontales (son adiabáticas).
2. Tenemos temperaturas TA y TB fijas en los extremos.
3. No se consideran fuentes ni sumideros.
4. $\kappa$ representa la conductividad térmica.
5. En este problema se desea calcular la temperatura T al interior del dominio.
6. Como primera aproximación el dominio se considera unidimensional. 



### Modelo matemático

El módelo matemático para este problema con $\kappa $ = constante es:

$-\kappa \frac{d^2T}{dx^2}=S$ ...(1)

$T(x=0)=TA$

$T(X=L)=TB$ 

Donde:
$\kappa$ = conductividad térmica [W/mK]

T = temperatura [K]

x = dominio o longitud [m]

S = Fuentes o sumideros

TA = Condicion de frontera de lado izquierdo

TB = Condición de frontera de lado derecho

### Modelo Numérico
    
2. Discretización de las ecuaciones
   
   Discretizar 1
   
   $\frac{d^2T}{dx^2}|_{i}=\frac{T_{i+1}-2T_i+T_{i-1}}{h^2}+O(h^2)$
   
   Donde $O(h^2)$ es el órden de la aproximación, que es de orden 2
   
   
   
   Sustiyuyendo en 1
   
   $-\kappa_i(\frac{T_{i+1}-2T_i+T_{i-1}}{h^2})=S_i$
   
   Si $r=\frac{k_i}{h^2}$, entonces (2):
   
   $-r_iT_i+1 + 2r_iT_i-r_iT_i-1 = S_i$
   
   Agregando las condiciones de Frontera: TA y TB
   
   Para i=1 $T_0=T_A$
   
   $-r_1T_0 + 2r_1T_1 - r_1T_2 = S_1$
   
   $-r_1T_A + 2r_1T_1 - r_1T_2 = S_1$
   
   $2r_1T_1 - r_1T_2 = S_1 + r_1T_A$
   
   Para i = N = 4 Tenemos $T_{N+1}=T_B$
   
   $-r_NT_{N-1} + 2r_NT_N - r_NT_{N+1} = S_N$
   
   $-r_NT_{N-1} + 2r_NT_N - r_NT_{B} = S_N$
   
   $-r_4T_{3} + 2r_4T_4  = S_4 + r_4T_{B}$
   
   El sistema completo:
   
   $2r_{1}T_{1}-r_{1}T_{2} = S_{1} + r_{1}T{A} $

   $-r_2T_1 + 2r_2T_2-r_2T_3 = S_2$

   $-r_3T_2 + 2r_3T_3-r_3T_4 = S_3$

   $-r_4T_3 + 2r_4T_4 = S_4 + r_4T_B$
   
   Lo anterior se puede expresar matricialmente como:
   
   $\begin{bmatrix}{2}&{-1}&{0}&{0}\\{-1}&{2}&{-1}&{0}\\{0}&{-1}&{2}&{-1}\\{0}&{0}&{-1}&{2}\end{bmatrix}
   \begin{bmatrix}{T_1}\\{T2}\\{T_3}\\{T_4}\end{bmatrix} = \frac{1}{r} \begin{bmatrix}{S_1}\\{S_1}\\{S_3}\\{S_4}\end{bmatrix} +        \begin{bmatrix}{T_A}\\{O}\\{0}\\{T_B}\end{bmatrix}$
   
   

### Modelo Computacional

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Parámetros para el estilo de las gráficas
# Si tienes instalado LaTeX, lo siguiente hace los textos 
# más estilizados. Si no tienes LaTeX, debes comentar la 
# línea: 'text.usetex': True
plt.style.use('seaborn-paper')
params = {'figure.figsize' : (14,7),
          'text.usetex'    : True,
          'xtick.labelsize': 20,
          'ytick.labelsize': 20,
          'axes.labelsize' : 24,
          'axes.titlesize' : 24,
          'legend.fontsize': 24,
          'lines.linewidth': 3,
          'lines.markersize': 10,
          'grid.color'     : 'darkgray',
          'grid.linewidth' : 0.5,
          'grid.linestyle' : '--',
          'font.family': 'DejaVu Serif',
         }
plt.rcParams.update(params)
```

```python
def buildMatrix(N):
    """
    Construye una matriz cuadrada de tamaño N X N
    
    El único parámetro de entrada que necesita es N, donde N es el número de renglones y columnas.
    
    
    """
    
    
    #Matriz de ceros
    A = np.zeros((N,N))
    
    #Primer renglón
    A[0,0] = 2
    A[0,1] = -1
    #Renglones interiores
    for i in range (1, N-1):
        A [i,i] = 2
        A [i,i+1] = -1
        A [i,i-1] = -1
    #Último renglón
    A [N-1,N-2] = -1
    A [N-1,N-1] = 2
       
       
    return A
```

```python
help(buildMatrix)
```

```python
#Parámetros físicos
L = 1.0
TA = 1
TB = 0
k = 1.0
S = 0.0
```

```python
#Parámetros numéricos
N = 4
h = L / (N+1)
r = k / h**2
```

```python
#Arreglo para almacenar la solución
T = np.zeros (N+2)
T[0]  = TA  #Frontera Izquierda
T[-1] = TB  #Frontera derecha
```

```python
#Lado derecho del sistema
b = np.zeros(N)
b[:] = S / r  #Fuente o Sumidero
b[0] += T[0]  #Condición de frontera
b[-1] += T[-1] #Condición de frontera
```

```python
#Construcción de la matriz
A = buildMatrix(N)
```

```python
#Solución del sistema lineal
C = np.linalg.inv(A)
T[1:N+1] = np.dot(C,b)
```

```python
#Impresión y graficación de la solución
print(T)
x = np.linspace(0, L, N+2)
plt.figure(figsize=(10,7))
plt.plot(x, T, c="grey", lw=2.0)
plt.scatter(x, T, edgecolors="k", zorder= 10)
plt.xlabel("$X$")
plt.ylabel("$T$")
plt.grid()
plt.show()
```

### Solución Analítica

$T(x)=(\frac{TB-TA}{L}+\frac{S}{2K}(L-x))x+TA$

### Análisis y explicación de la gráfica
Analizando la solución analítica 
Si S=0

$T(x)=(\frac{T_A-T_B}{L})x+TA$

Que es igual a
$y = mx + b$

Por lo que si no hay una fuente, la conducción de calor se hará de manera lineal. 

```python
#Código para la solución analítica:

def solExact(x, TA, TB, k, L, S):
    """
    Cálculo de la solución exacta
    
    Como parámetros de entrada necesita:
    x = vector que contiene los números para evaluar la función
    TA = Condición de frontera inicial
    TB = Condición de frontera final
    k = Conductividad térmica
    L = longitud
    S = Fuente o Sumidero
    
    
    """
    return ((TB-TA)/L+(S/(2*k))*(L-x))*x+TA
```

```python
help(solExact)
```

```python
def SolNum(L, N, k, S, A, b, T, etiqueta):
    """
    Calcula la solución numérica del problema
    
    Como parámetros de entrada necesita:
    L (Longitud), 
    N(número de grados de libertad o incógnitas),
    k (conductividad térmica), S(Fuente o Sumidero),
    A (matriz A o kernel del sistema), 
    b (vector columna que almacena la fuente y las condiciones de frontera), 
    T (vector columna que almacena la solucion) y 
    etiqueta para la gráfica.
    
    
    """
    
    
    
    h = L/ (N+1)
    r = k/ h**2
    
    
    # Lado derecho del sistema
    b = np.zeros(N) 
    b[:] =   S / r    #Fuente o sumidero
    b[0] +=   T[0]    # Condición de Frontera
    b[-1] += T[-1]    # Condición de Frontera
    
    #Solución del sistema lineal
    T[1:N+1] =np.linalg.solve(A,b)
    
    # Impresion y graficación de la solución
    x = np.linspace(0, L, N+2)
    
    #Impresión y graficación de la solución
    if etiqueta == "L":
        etiqueta = "$L$ = {:3.2f}".format(L)
    elif etiqueta == "k":
        etiqueta = "$\kappa$ = {:3.2f}".format(k)
    elif etiqueta == "S":
        etiqueta = "$S$ = {:3.2f}".format(S)
    
    #Se grafican los puntos de la solución
    plt.scatter(x, T, edgecolor="k", s=50, zorder=10, label=etiqueta)
```

```python
help(SolNum)
```

```python
def plotSol(title):
    
    """
    Define los parámetros para graficar la solución
    
    Como parámetro de entrada, solo necesita el título.
    
    """
    
    plt.suptitle("Conducción Estacionaria", fontsize=24, y=0.94, va="center_baseline")
    plt.title(title, fontsize=20, color="blue")
    plt.ylabel("$T$")
    plt.xlabel("$x$")
    plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), fontsize=12)
    plt.grid()
    plt.show()
```

### Análisis y explicación de la gráfica

Analizando nuevamente la solución analítica

$T(x)=(\frac{TB-TA}{L}+\frac{S}{2K}(L-x))x+TA$

Para todos los parámetros $\neq$ de 0

$T(x) = (T_B-T_A)(\frac{x}{L}) + \frac{SLx}{2k} - \frac{Sx^2}{2k} + T_A $

Debido a que tenemos términos cuadráticos, se puede observar ese comportamiento parabólico.


Por otro lado, hay dos términos que dependen de L y S

$(\frac{SLx}{2k})$......(Eq1)

$-\frac{Sx^2}{2k}$......(Eq2)

Si  $L_{--->0}$ 

Eq 1 $\approx$ 0 por lo que la fuente no va a afectar a toda la longitud y de acuerdo a lo que se observa en Eq 2, solo va a afectar a un punto en específico.


Por otro lado si $L_{--->\infty}$

La fuente va a contribuir tanto a toda la longitud L como en un punto en específico x. 

### Variando la Conductividad Térmica

```python
# Parámetros Físicos

L2 = 1.0
TA = 1.0
TB = 0.0
K = [0.1, 0.15, 0.25, 0.5, 1.0, 2.0, 10]
S2 = 1.0

# Parametros numéricos
N = 10

# Arreglo para almacenar la solución
T2 = np.zeros(N+2)
T2[0]  = TA # Frontera Izquierda
T2[-1] = TB #Frontera Derecha

# Contrucción de la matriz
A2 = buildMatrix(N)

for k in K:
    SolNum(L2, N, k, S2, A2, b, T2, "k")
    xe = np.linspace(0,L2,100)
    plt.plot(xe, solExact(xe, TA, TB, k, L2, S2), "k-", lw=1.0, alpha=0.5)
        
plotSol( "$L$ = {:3.2f}".format(L2) + "$S$ = {:3.2f}".format(S1))
```

### Análisis y explicación de la gráfica

Analizando la solución analítica

$T(x)=(\frac{TB-TA}{L}+\frac{S}{2K}(L-x))x+TA$

Podemos observar que si $k_{--->\infty}$

Entonces la ecuación anterior elimina el término cuadrático, comportandose únicamente como una recta.

Por otro lado si $k_{--->0}$ se conserva todos los términos y se comporta de manera parabólica.





### Variando la Fuente

```python
# Parámetros Físicos

L3 = 1.0
TA = 1.0
TB = 0.0
k3 = 1.0
s = [-6.0, -4.0, -2.0, 0.0, 2.0, 4.0, 6.0]

# Parametros numéricos
N = 10

# Arreglo para almacenar la solución
T3 = np.zeros(N+2)
T3[0]  = TA # Frontera Izquierda
T3[-1] = TB #Frontera Derecha

# Contrucción de la matriz
A2 = buildMatrix(N)

for S in s:
    SolNum(L3, N, k3, S, A2, b, T3, "S")
    xe = np.linspace(0,L3,100)
    plt.plot(xe, solExact(xe, TA, TB, k3, L3, S), "k-", lw=1.0, alpha=0.5)
        
plotSol( "$L$ = {:3.2f}".format(L3) + "$\kappa$ = {:3.2f}".format(k3))
```

### Análisis y explicación de la gráfica

Tal como habíamos comentado previamente, si analizamos la solución analítica

$T(x)=(\frac{TB-TA}{L}+\frac{S}{2K}(L-x))x+TA$

Si la fuente es 0, entonces el comportamiento de la ecuación será el de una línea recta.

Si la fuente va incrementando su valor, entonces tendremos un aumento en la temperatura, reflejado con una curva positiva.
Mientras que si se trata de un sumidero (perdidad de calor), la gráfica refleja la périda de energía con una curva negativa.
