# Modelo matemático

### Caso Estacionario

$ 0 = D_{th} \frac{\partial T^2}{\partial z^2}$

$D_{th} = \frac{\kappa}{\rho c}$

Donde:
1. $D_{th}$ = Difusividad térmica [$10^{-7} \frac{m^2}{s}$]
2. $T$ = Temperatura [°C]
3. $z$ = Profundidad [m]
4. $\kappa$ = Conductividad térmica [$\frac{W}{m  ^°C}$]
5. $\rho$ = Densidad de masa [$\frac{kg}{m^3}$]
6. $c$ = capcaidad térmica específica [$\frac{J}{kg K}$]


### Caso no estacionario

$ \frac{\partial T}{\partial t} = D_{th} \frac{\partial T^2}{\partial z^2}$

$D_{th} = \frac{\kappa}{\rho c}$

Donde: 

1. $D_{th}$ = Difusividad térmica [$10^{-7} \frac{m^2}{s}$]
2. $T$ = Temperatura [°C]
3. $z$ = Profundidad [m]
4. $t$ = Tiempo [años]
5. $\kappa$ = Conductividad térmica [$\frac{W}{m  ^°C}$]
6. $\rho$ = Densidad de masa [$\frac{kg}{m^3}$]
7. $c$ = capcaidad térmica específica [$\frac{J}{kg K}$]

# Modelo Computacional, caso estacionario

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fdm_1D import DTH, calcDth, harmonicMean, arithmeticMean, Laplaciano1D, interpTemp, Laplaciano1D_NS

# Parámetros para el estilo de las gráficas
plt.style.use('seaborn-paper')
params = {'figure.figsize' : (14,7),
          'text.usetex'    : True,
          'xtick.labelsize': 20,
          'ytick.labelsize': 20,
          'axes.labelsize' : 24,
          'axes.titlesize' : 24,
          'legend.fontsize': 24,
          'lines.linewidth': 3,
          'lines.markersize': 10,
          'grid.color'     : 'darkgray',
          'grid.linewidth' : 0.5,
          'grid.linestyle' : '--',
          'font.family': 'DejaVu Serif',
         }
plt.rcParams.update(params)
```

```python
# Parámetros Físicos del Problema

Ttop = 15
Tbot = 750
L = 4000
Depth1 = np.array([0, 50, 250, 400, 600, 800, 1000, 1500, 1900])
Depth2 = np.array([50, 250, 400, 600, 800, 1000, 1500, 1900, 4000])
k = np.array([1.84, 1.41, 1.36, 1.44, 1.46, 1.68, 1.84, 1.66, 2.10])
Rho = np.array([2120, 2340, 2300, 2300, 2290, 2700, 2700, 2360, 2760])
c_k = np.array([840, 840, 840, 840,  840, 840, 890, 890, 950])

Datos = pd.DataFrame(np.array([ Depth1, Depth2, k, Rho, c_k]).transpose() ,  
                    columns = [ "Depth1", "Depth2", "k", "Rho", "c_k"])

Datos["Difusividad"] = k / (Rho * c_k)

print('\nDatos del problema \n{}'.format(Datos))
```

```python
# Datos para la solución numérica por diferencias finitas
#
N = 99                      # Número de incógnitas
h = L / (N+1)              # Distancia entre nodos
z = np.linspace(0, L, N+2)  # Coordenada z de la profundidad

print('\nDatos para la solución numérica usando FD')
print(' dx = {:10.1f} [m] \n L  = {:10.1f} [m] \n N  = {:8d}'.format(h, L, N))
print('\nz shape = {} \n{}'.format(z.shape,z))
```

```python
Dth_data = Datos.Difusividad.to_numpy()
Dth = calcDth(Dth_data, z, N)

plt.plot((Dth/1e-7),z)
plt.gca().invert_yaxis()
plt.title('Difusividad Térmica vs Temperatura')
plt.xlabel('$Dth$ $10^{-7} [m^2/s]$')
plt.ylabel('Depth [m]')
plt.minorticks_on()
plt.grid()
```

```python
# SOLUCIÓN USANDO DIFERENCIAS FINITAS

# Definición del sistema lineal de N x N
gamma = Dth / h**2
f = np.zeros(N)            # Lado derecho del sistema
A = Laplaciano1D(N, gamma) # Matriz del sistema

# Arreglo para la solución
T = np.zeros(N+2)

# Aplicación de las condiciones de frontera tipo Dirichlet.
#
T[0] = Ttop
T[N+1] = Tbot 
f[0] -= Ttop * gamma[0]
f[N-1] -= Tbot * gamma[N+1]
# Solución del sistema lineal
T[1:N+1] = np.linalg.solve(A,f)


# GRAFICACIÓN
#
ax1 = plt.subplot(121)
plt.plot(T,z, 'r-', lw=2.0, label='FDM', zorder=10)
plt.plot([T[0], T[-1]], [z[0], z[-1]], '--', lw=0.75, c='k' )
plt.ylim(4100,-100)
plt.xlim(-10,800)
plt.title('Solución Numérica')
plt.xlabel('$T$ [$^o$C]')
plt.ylabel('Depth [m]')
plt.minorticks_on()
plt.legend()
plt.grid()

ax2 = plt.subplot(122, sharey=ax1)
plt.plot(Dth, z, '.--', lw=0.75, c='k')
plt.xlabel('$D_{th}$ [m$^2$/s]')
plt.xlim(0.65e-6,1.05e-6)
plt.minorticks_on()
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
plt.title('Coeficiente $D_{th}$')
plt.grid()

plt.subplots_adjust(hspace=1.0)

plt.show()
```

# Modelo computacional, caso no estacionario

```python
# Datos para la solución numérica por diferencias finitas
#
N = 99                      # Número de incógnitas
dx = L / (N+1)              # Distancia entre nodos
z = np.linspace(0, L, N+2)  # Coordenada z de la profundidad
dt = 3600*24*365 # Paso de tiempo (seconds per year)
NT = 250000      # Número total de pasos de tiempo

print('\nDatos para la solución numérica usando FD')
print(' dx = {:10.1f} [m] \n L  = {:10.1f} [m] \n N  = {:8d}'.format(dx, L, N))
print(' dt = {:10.1f} [s] \n NT = {:8d}'.format(dt, NT))
```

```python
# Aceleración para el método implícito
#
SP = 1000
dt *= SP
NT /= SP
print('\nUsamos un factor de aceleración debido al método implícito: ')
print(' SP = {:13d} (factor)\n dt = {:15.1f} [s] \n NT = {:13d} (pasos totales)'.format(SP, dt, int(NT)))
```

```python
# INTERPOLACIÓN DE LAS TEMPERATURAS

# Condición inicial: Profundidades
z_dat = [0, 100, 200, 400, 710, 803, 1100, 1200, 1400, 1500, 1600, 1700, 1800, 2000, 2500, 3000, 3500, 4000]
# Condición inicial: Temperaturas
T_dat = [15, 113, 145, 178, 155, 201, 215, 282, 223, 226, 252, 284, 310, 350, 450, 550, 650, 750]

# Interpolación a los lugares de la simulación
T_ini = interpTemp(z_dat, T_dat, z)

# Temperaturas actuales
z_dat_now = [0, 96, 192, 389, 596, 798, 1000, 1197, 1408, 1495, 1586, 1649, 1702, 1750, 1803, 1851, 1904, 1942, 1967, 2000, 2500, 3000, 3500, 4000]
T_dat_now = [15, 69, 74, 95, 121, 159, 168, 192, 213, 213, 236, 251, 267, 272, 284, 288, 298, 293, 310, 325, 440, 550, 650, 750]

# Interpolación a los lugares de la simulación
T_final = interpTemp(z_dat_now, T_dat_now, z)
```

```python
# SOLUCIÓN USANDO DIFERENCIAS FINITAS

# Definicion del sistema lineal de N x N
dt_dx2 = dt / dx**2
gamma = Dth * dt_dx2
A = Laplaciano1D_NS(N, gamma) # Matriz del sistema

#print('\nGamma shape = {}, A shape = {}'.format(gamma.shape, A.shape))

# Ciclo de solución temporal
T_new = T_ini.copy()
T_now = np.zeros(N+2)
for i in range(0, int(NT)+1):
    #print(i)
    T_new[ 1] += gamma[0] * T_ini[0]
    T_new[-2] += gamma[-1] * T_ini[-1]    
    T_new[1:-1] = np.linalg.solve(A,T_new[1:-1])
    if (i == 7):
        T_now = T_new.copy() # Temperatura actual

# GRAFICACIÓN
ax1 = plt.subplot(121)
plt.plot([T_ini[0], T_ini[-1]], [z[0], z[-1]], '--', lw=0.75, c='k' )
plt.scatter(T_dat, z_dat, s=20, alpha=0.5)
plt.plot(T_ini, z, '-', lw = 1.5, label='Cond. inicial')
plt.scatter(T_dat_now, z_dat_now, s=20, alpha=0.5, label='Datos de pozo')
plt.plot(T_now, z, label='Simulación Temp. actual')
plt.plot(T_new,z, label='Simulación Temp. final')
plt.ylim(4100,-100)
plt.xlim(-20,800)
plt.title('Solución Numérica')
plt.xlabel('$T$ [$^o$C]')
plt.ylabel('Depth [m]')
plt.grid()
plt.legend()

ax2 = plt.subplot(122, sharey=ax1)
plt.plot(Dth, z, 'k.--', lw=0.75)
plt.xlabel('$D_{th}$ [m$^2$/s]')
plt.xlim(0.65e-6,1.05e-6)
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
plt.title('Coeficiente $D_{th}$')
plt.grid()

plt.subplots_adjust(hspace=1.0)

plt.show()
```

```python
# Datos para la solución numérica por diferencias finitas
#
Na = 115                      # Número de incógnitas
dxa = L / (Na+1)              # Distancia entre nodos
za = np.linspace(0, L, Na+2)  # Coordenada z de la profundidad
dta = 3600*24*365 # Paso de tiempo (seconds per year)
NTa = 400000      # Número total de pasos de tiempo
# Aceleración para el método implícito
#
SPa = 1000
dta *= SP
NTa /= SP
T_inia = interpTemp(z_dat, T_dat, za)
T_finala = interpTemp(z_dat_now, T_dat_now, za)
Dtha = calcDth(Dth_data, za, Na)
# SOLUCIÓN USANDO DIFERENCIAS FINITAS

# Definicion del sistema lineal de N x N
dt_dx2a = dta / dxa**2
gammaa = Dtha * dt_dx2a
Aa = Laplaciano1D_NS(Na, gammaa) # Matriz del sistema

#print('\nGamma shape = {}, A shape = {}'.format(gamma.shape, A.shape))

# Ciclo de solución temporal
T_newa = T_inia.copy()
T_nowa = np.zeros(Na+2)
for i in range(0, int(NTa)+1):
    #print(i)
    T_newa[ 1] += gammaa[0] * T_inia[0]
    T_newa[-2] += gammaa[-1] * T_inia[-1]    
    T_newa[1:-1] = np.linalg.solve(Aa,T_newa[1:-1])
    if (i == 7):
        T_nowa = T_newa.copy() # Temperatura actual

# GRAFICACIÓN
ax1 = plt.subplot(121)
plt.plot([T_ini[0], T_ini[-1]], [z[0], z[-1]], '--', lw=0.75, c='k' )
#plt.scatter(T_dat, z_dat, s=20, alpha=0.5)
plt.scatter(T_dat_now, z_dat_now, s=20, alpha=0.5, label='Datos de pozo')
plt.plot(T_now, z, label='Temp. actual N= 99')
plt.plot(T_new,z, label='Temp. final N=99')
plt.plot(T_nowa, za, label='Temp. actual N =')
plt.plot(T_newa,za, label='Temp. final N=')
plt.ylim(4100,-100)
plt.xlim(-20,800)
plt.title('Solución Numérica')
plt.xlabel('$T$ [$^o$C]')
plt.ylabel('Depth [m]')
plt.grid()
plt.legend(fontsize=9)

ax2 = plt.subplot(122, sharey=ax1)
plt.plot(Dth, z, 'k.--', lw=0.75)
plt.xlabel('$D_{th}$ [m$^2$/s]')
plt.xlim(0.65e-6,1.05e-6)
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
plt.title('Coeficiente $D_{th}$')
plt.grid()

plt.subplots_adjust(hspace=1.0)

plt.show()
```
